```julia
print("Input lamda: \n")
l = parse(Float64, readline())
X=[]
for i in 0:10000
    u=rand()
    j=0
    p=exp(-l)
    A=p
    
    while true
        if u<A 
            push!(X,j)
            break
        else
            p=(l/(j+1))*p
            A+=p
            j+=1
        end    
    end
end
using Plots
histogram(X)
```
