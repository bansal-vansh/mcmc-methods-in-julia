# PROBLEM 1

### Implement a Discrete Inverse Transform sampler for Poisson distribution.

```julia
# Below is the function which stimulates a poisson Random Variable and can be used to get a Random Whole Number which follows
# poisson's probability distribution.

function poisson(lambda::Float64)
    u = rand();
    sum = 1;
    pro1 = 1;
    pro2 = 1;
    ctr = 0;
    con = (MathConstants.e)^(-1*lambda);
    while (con*sum) <= u
        pro1=pro1*lambda;
        pro2=pro2*(ctr+1);
        ctr=ctr+1;
        sum=sum+(pro1/pro2);
    end
    return ctr;
end

# Function which returns an array of required size containing a set of Independent and Identically distributed Poisson
# Random Variables of the required parameter lambda.

function simulate1(lambda::Float64,length::Integer)
    arr = [];
    ctr = 0;
    while ctr<length
        c = poisson(lambda);
        push!(arr,c);
        ctr = ctr+1;
    end
    return arr;
end
```

```julia
# Theorectically the mean of a Poisson(lambda) Random Variable = lambda
# Theorectically the variance of a Poisson(lambda) Random Variable = lambda

lambda = readline();
lambda = parse(Float64,lambda);
```

```julia
simulate1(lambda,30000) # An example implementation showing the samples
```

```julia
display(plot(histogram(simulate1(lambda,30000))))
```

```julia
display(plot(simulate1(lambda,30000)))
```

```julia
mean(simulate1(lambda,30000)) # Comes out to be almost equal to chosen lambda
```

```julia
var(simulate1(lambda,30000)) # Comes out to be almost equal to chosen lambda
```

# PROBLEM 2

### Implement a Discrete Accept Reject Sampler to simulate draws from $Binomial(n, p)$ using a Poisson proposal.

```julia
# Poission Distribution sampler in the above question would work as the proposal distribution for using the Accept Reject Method 
# of sampling of a Binomial(n,p) Random Variable.

# The main sampler function which uses the Accept Reject Sampling technique for simulating a Binomial(n,p) Random Variable
# with the Poisson Distribution as a proposal. The proposal is being created using the Inverse Transform Sampler.
# CONSTRAINT : The value of 64 Bit Floating point number p should follow -- 0 < p < 1

function binomial(n::Integer,p::Float64)
    flag = true;
    c = ((MathConstants.e)^(p))*factorial(n);
    y = 0;
    while flag == true
        u = rand();
        y = poisson(p);
        q_y = ((MathConstants.e)^(-1*p))*(p^y)/factorial(y);
        p_y = 0;
        if y <= n
            p_y = (p^y)*((1-p)^(n-y))*factorial(n)/(factorial(y)*factorial(n-y));
        else 
            p_y = 0;
        end
        num = p_y/(c*(q_y));
        if u < num
            flag = false;
        else
            continue;
        end 
    end
    return y;
end

# Function which returns an array of required size containing a set of Independent and Identically distributed Binomial(n,p)
# Random Variables.

function simulate2(n::Integer,p::Float64,length::Integer)
    arr = [];
    ctr = 0;
    while ctr<length
        c = binomial(n,p);
        push!(arr,c);
        ctr = ctr+1;
    end
    return arr;
end
```

```julia
# Theorectically the mean of a Binomial(n,p) Random Variable = n*p
# Theorectically the variance of a Binomial(n,p) Random Variable = n*p*q

n = readline();
n = parse(Int64,n);
p = readline();
p = parse(Float64,p);
```

```julia
simulate2(n,p,30000) # An example implementation
```

```julia
display(plot(histogram(simulate2(n,p,30000))))
```

```julia
display(plot(simulate2(n,p,30000)))
```

```julia
mean(simulate2(n,p,30000)) # Comes almost equal to n*p for large sample size
```

```julia
var(simulate2(n,p,30000)) # Comes almost equal to n*p*q where q = 1-p for large sample size
```
