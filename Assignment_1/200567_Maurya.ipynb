```julia
print("Input lamda: \n")
l = parse(Float64, readline())
X=[]
for i in 0:10000
    u=rand()
    j=0
    p=exp(-l)
    A=p
    
    while true
        if u<A 
            push!(X,j)
            break
        else
            p=(l/(j+1))*p
            A+=p
            j+=1
        end    
    end
end
using Plots
histogram(X)
```

```julia
print("Input n: \n")
n = parse(Float64, readline())
print("Input p: \n")
p = parse(Float64, readline())
X=[]
m=0
for x in 0:n
    c=binomial(Int64(n),Int64(x))*((1-p)^(n-x))*exp(p)/factorial(big(Int64(x)))
    if c>m
        m=c
    end
end
m+=0.01
for i in 0:10000
    a=0
    while a==0
        u=rand()
        j=rand((0:n))
        r=binomial(Int64(n),Int64(j))*((1-p)^(n-j))*exp(p)/(factorial(big(Int64(j)))*m)
        if u<r
            a=1
            push!(X,j)
        end
    end
end
using Plots
histogram(X)
```
