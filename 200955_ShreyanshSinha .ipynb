```julia
using Distributions
using Random
```

```julia
function poisson(λ)
    U = rand(Uniform(0,1))
    i = 0
    p = exp(-λ)
    A = p
    X = 0
    while true
        if U<A
            X = i
            break
        else 
            p = λ*(1/(1+i))*p 
            A = A + p
            i = i + 1
        end
    end
    
    return X
end
```
