# QUESTION-1

```julia
using Distributions
```

$$c_x = 2, p_x = 0.4, c_y = 1, p_y = 0.6$$

```julia
iter = 10000
C = Bernoulli(0.43)
target = rand(C, iter)
using Plots
histogram(target)
```

```julia
Prop = Int64[]
counting = Int64[]
for i in 1:iter
    accept=0
    count=0
    while (accept==0)
        count=count+1
        C1 = Bernoulli(0.33)
        D1 = rand(C1, 1)
            if (D1[1]==1)
                C2 = Bernoulli(0.6)
                D2 = rand(C2, 1)
                if (D2[1]==1)
                    push!(Prop, 1)
                    push!(counting, count)
                    accept=1
                end
            end
            if (D1[1]==0)
                C3 = Bernoulli(0.4)
                D3 = rand(C3, 1)
                if (D3[1]==1)
                    push!(Prop, 0)
                    push!(counting, count)
                    accept=1
                end
            end
    end
end
```

```julia
histogram(Prop)
```

```julia
mean(counting)
```

Since the Histogram for target and Prop are same, then it is proved that given algorithm generates a bernoulli($\tfrac{c_y*p_y}{c_x*p_x + c_y*p_y})$

# QUESTION-2

uniform distribution p-dimensional hypercube =$ 1/2^p$

So for p=2,
Target distribution, $f(x,y) = \tfrac{I(x^2+y^2<1)}{\pi}$

Proposal distribution, $g(x,y) = \tfrac{1}{4}$

$c = \tfrac{4}{\pi} > \tfrac{f(x,y)}{g(x,y)}  $

therefore we have $\tfrac{f(x,y)}{c*g(x,y)} = I(x^2+y^2<1)$

```julia
value1=Float64[]
value2=Float64[]
```

```julia
for i in 1:10000
    count=0
    accept=0
    while(accept==0)
        
        prop1=rand(Uniform(-1,1))
        prop2=rand(Uniform(-1,1))
        euclid_dist = prop1*prop1+prop2*prop2
        if (euclid_dist<1)
            push!(value1,prop1)
            push!(value2,prop2)
            
            accept=1
        end
    end
end
```

```julia
plot(value1,value2)
```

For p = 3
Target distribution, $f(x,y,z) = \tfrac{(\tfrac{3\pi^{0.5}}{4}I(x^2+y^2+z^2<1)}{\pi^{1.5}}$
                                $=  \tfrac{3*I(x^2+y^2+z^2<1)}{4*\pi} $

Proposal distribution for p=3, g(x,y,z) = $\tfrac{1}{8}$

$ c = \tfrac{6}{\pi} < \tfrac{f(x,y,z)}{g(x,y,z)} $

therefore we have $\tfrac{f(x,y)}{c*g(x,y)} = I(x^2+y^2+z^2<1)$

```julia
value1=Float64[]
value2=Float64[]
value3=Float64[]
```

```julia
for i in 1:10000
    count=0
    accept=0
    while(accept==0)
        
        prop1=rand(Uniform(-1,1))
        prop2=rand(Uniform(-1,1))
        prop3=rand(Uniform(-1,1))
        euclid_dist = prop1*prop1+prop2*prop2+prop3*prop3
        if (euclid_dist<1)
            push!(value1,prop1)
            push!(value2,prop2)
            push!(value3,prop3)
            accept=1
        end
    end
end
```

```julia
plot(value1,value2,value3)
```
